```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the processed data
file_path = "processed_data.csv"
df = pd.read_csv(file_path)
```

```python

# Step 2: Display first few rows to verify data integrity
print("First five rows of the processed dataset:")
print(df.head())
```

```python

# Step 3: Summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python

# Step 4: Visualizing the distribution of Grip Strength
plt.figure(figsize=(8,5))
sns.histplot(df['Grip_strength'], bins=5, kde=True)
plt.title("Distribution of Grip Strength")
plt.xlabel("Grip Strength (kg)")
plt.ylabel("Frequency")
plt.show()
```

```python

# Step 5: Boxplot for Grip Strength by Frailty
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Frailty'], y=df['Grip_strength'])
plt.title("Grip Strength vs. Frailty")
plt.xlabel("Frailty")
plt.ylabel("Grip Strength (kg)")
plt.show()
```

```python

# Step 6: Scatter plot - Age vs Grip Strength
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Age'], y=df['Grip_strength'], hue=df['Frailty'], palette='coolwarm')
plt.title("Age vs Grip Strength")
plt.xlabel("Age (years)")
plt.ylabel("Grip Strength (kg)")
plt.show()
```

```python

# Step 7: Correlation heatmap to identify relationships between variables
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python

# Step 8: Bar Plot - Average Grip Strength by Frailty
plt.figure(figsize=(8,5))
df.groupby("Frailty")["Grip_strength"].mean().plot(kind='bar', color=['blue', 'red'])
plt.title("Average Grip Strength by Frailty")
plt.xlabel("Frailty (0 = No, 1 = Yes)")
plt.ylabel("Average Grip Strength (kg)")
plt.show()

print("Analysis complete.")
```
