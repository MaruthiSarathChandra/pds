```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the student performance dataset
file_path = "processed_student_data.csv"
df = pd.read_csv(file_path)
```

```python
# Display first few rows to verify data integrity
print("First five rows of the dataset:")
print(df.head())
```

```python
# Step 1: Histogram of Math Scores
plt.figure(figsize=(8,5))
sns.histplot(df['math score'], bins=10, kde=True, color='blue')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Step 2: Boxplot of Scores by Gender
plt.figure(figsize=(8,5))
sns.boxplot(x=df['gender'], y=df['math score'], palette='coolwarm')
plt.title("Math Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=df['gender'], y=df['reading score'], palette='coolwarm')
plt.title("Reading Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Reading Score")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x=df['gender'], y=df['writing score'], palette='coolwarm')
plt.title("Writing Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Writing Score")
plt.show()
```

```python
# Step 3: Bar Plot - Impact of Test Preparation Course
plt.figure(figsize=(8,5))
df.groupby("test preparation course")[['math score', 'reading score', 'writing score']].mean().plot(kind='bar', figsize=(8,5), rot=0)
plt.title("Average Scores by Test Preparation Course Completion")
plt.xlabel("Test Preparation Course")
plt.ylabel("Average Score")
plt.legend(["Math", "Reading", "Writing"])
plt.show()
```

```python
# Step 4: Scatter Plot - Correlation Between Math and Reading Scores
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['math score'], y=df['reading score'], hue=df['gender'], palette='coolwarm')
plt.title("Math vs Reading Scores")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
# Step 5: Heatmap - Correlation Between All Scores
plt.figure(figsize=(8,5))
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Between Scores")
plt.show()

print("Task 2 Analysis Complete.")
```
